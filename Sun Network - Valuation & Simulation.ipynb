SUN TV VALUATION - 2022

```python
print('--PART 1 : Forecast Inputs--')
tI = float(input('Enter the Total Income (in crore INR): ')) # 3749 Total Income :1
ebit = float(input('Enter the current EBIT (in crore INR): '))# 2193 (Profit before tax):2
ec = float(input('Enter the net equity capital (in crore INR): ')) # 197 Total Equity Capital :3

n = int(input('Enter the number of years: ')) # 5
t = float(input('Enter the effective tax rate (in %): ')) #25 % 
```

```python
cI = float(input('Enter the re-invested cash (in crore INR): ')) # 970 :4
dinc =  float(input('Enter the dividend income (in crore INR): ')) # 116 : 5
capex = float(input('Enter the net CAPEX (in crore INR): ')) # 595 - 6a
dWC = float(input('Enter the change in working capital (in crore INR): ')) # 150 - 6b
```

```python
rI = (cI/tI)*100
r = rI * (dinc/ec)
print('The re-investment rate is ', round(rI,2), '%')
print('The growth rate is ', round(r,2), '%')
```

```python
print('--PART 1 - FREE CASH FLOW --')
cf = (ebit*(1-t/100)-capex-dWC)

print('Cash Flow for current year is ', round(cf), 'crore INR')
rein = ebit*(1-t/100) - cf
print('Re-investment for current year is ', round(rein), 'crore INR')


print('----------------------')
print('--Forecast EBIT--')
e1 = ebit
e = e1
print('EBIT for current year is', round(e,2), 'crore INR')
for i in range(0, n):
    e = ebit*(1+r/100)
    ebit=e
    print('EBIT for forecast year no ', i+1, 'is', round(e,2), 'crore INR')

print('----------------------')
print('--Forecast EBIT(1-t)--')
e1t = e1*(1-t/100)
for j in range(0, n):
    e2t = e1t*(1+r/100)
    e1t=e2t
    print('EBIT(1-t) for forecast year no ', j+1, 'is', round(e1t,2), 'crore INR')
    print('RE-investment for forecast year no ', j+1, 'is', round((rI/100)*e1t,2), 'crore INR')
    
print('----------------------')
print('--Forecast FCFF--')
ef1=e1*(1-t/100)
L=[]
for k in range(0, n):
    ef2 = ef1*(1+r/100)
    ef1=ef2
    fcff=round(ef1-ef1*rI/100,2)
    L.append(fcff)
   
    print('FCFF for forecast year no ', k+1, 'is', round(ef1-ef1*rI/100,2), 'crore INR')
```

```python
print('--PART 2 - COST OF CAPITAL --')

g = float(input('Enter the risk free rate (%): ')) # 7.3% 10-Y GB
rp = float(input('Enter the risk premium (%): ')) # 8% 10-Y GB
beta = float(input('Enter the beta (estimated) : ')) #0.7

coe = g + beta*(rp)
coc = coe # Debt is hardly 10 %

print('Current Cost of Equity is ', round(coe,2), '%')
print('Current Cost of Capital is ', round(coc,2), '%')
```

```python
print('--Forecast CoC--')
cc1=1+(coc/100)
print('Cumulated CoC for forecast year no 1', round(cc1,2))
C=[]
C.append(cc1)
for i in range(1,n):
    cc2=cc1*(1+coc/100)
    cc1=cc2
    C.append(cc1)
    print('Cumulated CoC for forecast year no ', i+1, 'is', round(cc1,2))
    
P=[]
for i in range(0,n):
    pv=round(L[i]/C[i],2)
    P.append(pv)
PV1 = P[0]
for j in range(1,n):
    PV2 = PV1 + P[n-j] 
    PV1=PV2
print('The net present value (NPV) is ', round(PV2,2),'crore INR') 
```

```python
print('--PART 3 - INTRINSIC SHARE VALUE--')
ns = int(input('Enter the number of outstanding shares (in crore): ')) # 40 
rIg = g/coc * 100 # Assume steady flow - ROC ~ CoC

print('Reinvestment rate in stable growth is ', round(rIg,2),'%')
tt = float(input('Enter tax rate for terminal year (%): ')) # 25

print('----------------------')
cft = (e*(1-tt/100)*(1-rIg/100)*(1+g/100)) # Formula 2 from FCFE -> NI *(1+g)*(1-rI)
print('Expected FCFE (Terminal Value) ', round(cft,2),'crore INR')

TV = cft/((coc/100)-(g/100)) # Formula 3 from FCFE 
print('Terminal Value at end of year 5 is ', round(TV,2),'crore INR')

voe = TV / ((1+coc/100)**(n)) + PV2


print('Value of common equity in stock is ', round(voe,2),'crore INR')
print('Intrinsic value of equity per share is', round(voe/ns,2) , 'INR')
```

```python
voet = TV / ((1+coe/100)**(n+1)) 
```

```python
m = float(input('Enter the Market price (current value, in INR): ')) # Aug 2022 ~ 520
b = round(voe/ns,2)
print ('The Price to Book Ratio is ', round((m/b),2))
```

SIMULATION OF STOCK PRICE - FROM 2021 TO 2022 BY MONTE CARLO METHOD

```python
import numpy as np
import matplotlib.pyplot as plt


def geo_paths(S, T, r, q, sigma, steps, N):
    """
    Inputs
    #S = Current stock Price
    #K = Strike Price
    #T = Time to maturity 1 year = 1, 1 months = 1/12
    #r = risk free interest rate
    #q = dividend yield
    # sigma = volatility 
    
    Output
    # [steps,N] Matrix of asset paths 
    """
    dt = T/steps
    #S_{T} = ln(S_{0})+\int_{0}^T(\mu-\frac{\sigma^2}{2})dt+\int_{0}^T \sigma dW(t)
    ST = np.log(S) +  np.cumsum(((r - q - sigma**2/2)*dt +\
                              sigma*np.sqrt(dt) * \
                              np.random.normal(size=(steps,N))),axis=0)
    
    return np.exp(ST)

S = 500 #stock price S_{0}
K = 520 # strike
T = 1/12 # time to maturity (Monthly)
r = 0.07 # risk free risk in annual %
q = 0.01 # annual dividend rate
sigma = 0.25 # annual volatility in %
steps = 220 # time steps
N = 25 # number of trials

paths= geo_paths(S,T,r, q,sigma,steps,N)

plt.plot(paths);
plt.xlabel("Time Increments")
plt.ylabel("Stock Price")
plt.title("Sun TV Stock Price Simulation (2021-2022)")
```
